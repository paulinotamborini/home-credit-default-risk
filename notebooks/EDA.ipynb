## 🔍 Detección de Problemas en los Datos

### 1 Valores Faltantes (Missing Values)

```python
# Contar valores faltantes por columna
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

# Crear una tabla con los resultados
missing_df = pd.DataFrame({
    'Columna': missing.index,
    'Valores Faltantes': missing.values,
    'Porcentaje': missing_percent.values
})

# Mostrar solo columnas con valores faltantes
missing_df[missing_df['Valores Faltantes'] > 0].sort_values('Porcentaje', ascending=False)
```

```python
# Analizá si los nulos se distribuyen distinto entre clase 0 y clase 1
cols_with_nulls = df.columns[df.isnull().any()].tolist()

null_analysis = []
for col in cols_with_nulls:
    null_mask = df[col].isnull()
    default_rate_null = df.loc[null_mask, 'TARGET'].mean()
    default_rate_not_null = df.loc[~null_mask, 'TARGET'].mean()
    null_analysis.append({
        'feature': col,
        'pct_null': null_mask.mean(),
        'default_rate_when_null': default_rate_null,
        'default_rate_when_not_null': default_rate_not_null,
        'difference': default_rate_null - default_rate_not_null
    })

null_df = pd.DataFrame(null_analysis).sort_values('difference', ascending=False)
print(null_df.head(10))
print(null_df.tail(10))
```

## 📊 Análisis de la Variable Objetivo: Target

```python
# Contar cuántos Yes y No hay en Churn
target_counts = df['TARGET'].value_counts()
target_percent = df['TARGET'].value_counts(normalize=True) * 100

print('Distribución de TARGET:')
print('-' * 40)
for category, count in target_counts.items():
    percent = target_percent[category]
    print(f'{category:5} - {count:4} clientes ({percent:.1f}%)')

print('\n⚠️ IMPORTANTE: Este es un dataset DESBALANCEADO')
```

```python
# Visualizar la distribución con un gráfico de barras
plt.figure(figsize=(8, 6))
colors = ["#22a057", '#e74c3c']  # Verde para No, Rojo para Yes
bars = plt.bar(target_counts.index, target_counts.values, color=colors, alpha=0.7, edgecolor='black')

# Agregar los valores encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}\n({height/len(df)*100:.1f}%)',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Distribución de Target', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Target', fontsize=12)
plt.ylabel('Cantidad de Clientes', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 📈 6. Análisis de Variables Numéricas

Veamos cómo se distribuyen las variables numéricas.

```python
numeric_col = df.select_dtypes(include=['int64', "float64"]).columns
```

```python
# Identificar columnas numéricas
numeric_cols = ['AMT_INCOME_TOTAL', 'DAYS_BIRTH', "AMT_CREDIT"]

# Crear histogramas para cada variable
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, col in enumerate(numeric_cols):
    # Histograma
    axes[i].hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Distribución de {col}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Frecuencia', fontsize=12)
    axes[i].grid(axis='y', alpha=0.3)
    
    # Agregar línea vertical en la mediana
    median = df[col].median()
    axes[i].axvline(median, color='red', linestyle='--', linewidth=2, label=f'Mediana: {median:.1f}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
